```python
import numpy as np
import pandas as pd
import re

data = pd.read_csv('trump.xlsm - trump.csv')
```

```python
data.head()
```

```python
def clean_tweet(df):
    cleaned = df.text.replace('#', '') # Remove all '#' from the text
    cleaned = re.sub('http\S+', '', cleaned) # Remove all URLs from the text
    cleaned = re.sub('@\S+', '', cleaned) # Remove all @xxxx's from the text
    return cleaned
```

```python
cleaned_text = data.apply(clean_tweet, axis=1)
```

```python
data['text_cleaned'] = cleaned_text
data.head()
```

```python
data.to_csv('Trump Tweets Cleaned.csv', index=False)
```
